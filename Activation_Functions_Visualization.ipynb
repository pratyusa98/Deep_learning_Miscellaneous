```python
import matplotlib.pyplot as plt  # For plotting graphs
import numpy as np               # For exp, log etc.,
```

```python
z=np.arange(-6,6,0.01)
```

```python
z.shape
```

### Step Function

```python
def step(x):
    if x>=0:
        return 1
    else:
        return 0
```

```python
step_output = [step(i) for i in z]
fig, ax = plt.subplots(figsize=(9, 6))

ax.plot(z,step_output, linewidth=3, label="step")


ax.legend(loc="upper left", frameon=False)
ax.grid()
```

### Linear Function

```python
def linear(x):
    ''' y = f(x) It returns the input as it is'''
    return x
def der_linear(x):
    return 1
```

```python
step_output_l = [der_linear(i) for i in z]
fig, ax = plt.subplots(figsize=(9, 6))

ax.plot(z,linear(z), linewidth=3, label="linear")
ax.plot(z,step_output_l, linewidth=3, label="der_linear")

ax.legend(loc="upper left", frameon=False)
ax.grid()
```

### Sigmoid Function

```python
# Sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))

# Derivative of Sigmoid
def der_sigmoid(x):
    return sigmoid(x) * (1- sigmoid(x))
```

```python
fig, ax = plt.subplots(figsize=(9, 6))

ax.plot(z,sigmoid(z), linewidth=3, label="sigmoid")
ax.plot(z,der_sigmoid(z), linewidth=3, label="Derivative_sigmoid")


ax.legend(loc="upper left", frameon=False)
ax.grid()
```

### Tanh Function

```python
# Hyperbolic Tangent (htan) Activation Function
def tanh(x):
    return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

# htan derivative
def der_tanh(x):
    return 1 - tanh(x) * tanh(x)
```

```python
fig, ax = plt.subplots(figsize=(9, 6))

ax.plot(z,tanh(z), linewidth=3, label="tanh")
ax.plot(z,der_tanh(z), linewidth=3, label="Derivative_tanh")


ax.legend(loc="upper left", frameon=False)
ax.grid()
```

### Relu Activation Function

```python
# Rectified Linear Unit (ReLU)
def ReLU(x):
    data = [max(0,value) for value in x]
    return np.array(data, dtype=float)

# Derivative for ReLU
def der_ReLU(x):
    data = [1 if value>0 else 0 for value in x]
    return np.array(data, dtype=float)
```

```python
fig, ax = plt.subplots(figsize=(9, 6))

ax.plot(z,ReLU(z), linewidth=3, label="Relu")
ax.plot(z,der_ReLU(z), linewidth=3, label="Derivative_Relu")


ax.legend(loc="upper left", frameon=False)
ax.grid()
```

### Leaky Rectified Linear Unit (leaky ReLU) Function

```python
# Leaky Rectified Linear Unit (leaky ReLU) Activation Function
def leaky_ReLU(x):
    data = [max(0.05*value,value) for value in x]
    return np.array(data, dtype=float)

# Derivative for leaky ReLU 
def der_leaky_ReLU(x):
    data = [1 if value>0 else 0.05 for value in x]
    return np.array(data, dtype=float)
```

```python
fig, ax = plt.subplots(figsize=(9, 6))

ax.plot(z,leaky_ReLU(z), linewidth=3, label="leaky_relu")
ax.plot(z,der_leaky_ReLU(z), linewidth=3, label="Derivative_leaky_relu")


ax.legend(loc="upper left", frameon=False)
ax.grid()
```

### Softmax Function

```python
def softmax(x):
    ''' Compute softmax values for each sets of scores in x. '''
    return np.exp(x) / np.sum(np.exp(x), axis=0)
```

```python
fig, ax = plt.subplots(figsize=(9, 6))

ax.plot(z,softmax(z), linewidth=3, label="softmax")
# ax.plot(z,der_softplus(z), linewidth=3, label="Derivative_softplus")


ax.legend(loc="upper left", frameon=False)
ax.grid()
```

## All Activation Function

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(z,step_output, linewidth=3, label="step")
ax.plot(z,sigmoid(z), linewidth=3, label="sigmoid")
ax.plot(z,tanh(z), linewidth=3, label="tanh")
ax.plot(z,ReLU(z), linewidth=3, label="relu")
ax.plot(z,leaky_ReLU(z), linewidth=3, label="leaky_relu")
ax.plot(z,softmax(z), linewidth=3, label="softmax")

ax.grid()
ax.legend(loc="upper left", frameon=False)
```
